```python
import pandas as pd
import numpy as np

from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense
```

# Read the data from csv

```python
data = pd.read_csv('Mobile_Price_Classification-220531-204702.csv')
data.head(5)
```

# Split into features  (x)  and targets (y)

```python
x = data.drop('price_range', axis=1)
y = data['price_range']
```


# 3. Preprocess: Scale the features

```python
scaler = StandardScaler()
X_scaled = scaler.fit_transform(x)
```

# 4. Split into training and testing sets (75% train, 25% test)

```python
X_train, X_test, Y_train, Y_test = train_test_split(X_scaled, y, test_size=0.25, random_state=42)
```

# 5. Build the ANN model

```python
model = Sequential([
    Dense(8, activation='relu', input_shape=(X_train.shape[1],)),
    Dense(4, activation='relu'),
    Dense(1, activation='sigmoid')
])
```

# 6. Compile the model

```python
model.compile(optimizer='adam', loss='binary_crossentropy', metrics=['accuracy'])
```

# Step 7: Train the model

```python
model.fit(X_train, Y_train, epochs=100, batch_size=32, validation_split=0.2)
```


# Step 8: Save the weights

```python
model.save_weights('mobile_price_model_weights.weights.h5')
```
